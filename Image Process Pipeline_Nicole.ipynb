# Segmentation * (Saliency + Optical flow)


```
from skimage import feature
from sklearn.preprocessing import normalize
from sklearn.metrics.pairwise import euclidean_distances
import glob
```

### functions

```
def sal_mask(image):
  '''
  PARAMETERS:
    image: input image [h, w, 3]
  
  RETURN:
    origs: original images [h, w, 3, #images]
    origs_gray: original images in grayscale [h, w, #images]
    saliens: deepgaze saliency map of images [h, w, #images]
    masks: combined object masks of images [h, w, #images]
  '''
  h,w,_ = image.shape

  # get masks
  results = model.detect([image])
  r = results[0]
  obj_masks = r["masks"]
  # fig = visualize.display_instances(image, r['rois'], r['masks'], r['class_ids'], 
  #                           class_names, r['scores'])

  # get mask edges
  num_objs = r["masks"].shape[2]
  obj_edges = np.zeros(r["masks"].shape)
  for o in range(num_objs):
    m = r["masks"][:,:,o]
    edges = feature.canny(m, sigma=5).astype("uint8")
    kernel = np.ones((10,10), np.uint8) 
    edges_dilated = cv2.dilate(edges,kernel)
    obj_edges[:,:,o] = edges_dilated

  # # get all object masks into one mask
  # combined_masks = np.sum(r['masks'],2)

  # get saliency map
  my_map = FasaSaliencyMapping(h, w) 
  image_salient = my_map.returnMask(image, tot_bins=8, format='BGR2LAB')  # get the mask from the original image
  image_salient_blr = cv2.GaussianBlur(image_salient, (3,3), 1)

  return image_salient_blr, obj_masks, r["rois"], r["class_ids"], obj_edges #roi top-left and bottom-right y,x


def intersec_area(a, b):  # returns 0 if rectangles don't intersect
    dx = np.min([a[3], b[3]]) - np.max([a[1], b[1]])
    dy = np.mean([a[2], b[2]]) - np.max([a[0], b[0]])
    if (dx>=0) and (dy>=0):
        return dx*dy
    else: 
      return 0
```

### read original video and set up

```
vid_path = "/content/gdrive/My Drive/291 Bionic Vision Project/Asa_Video/stim6_vid(done)" # ORIGINAL VIDEO PATH FOLDER 
vid_name = 'stim6.mp4'

cap = cv2.VideoCapture('%s/%s'% (vid_path, vid_name))
ret, frame1 = cap.read()

results = model.detect([frame1])
rs = results[0]
# fig = visualize.display_instances(frame1, rs['rois'], rs['masks'], rs['class_ids'], 
                          # class_names, rs['scores'])


# set up optical flow 
prvs = cv2.cvtColor(frame1,cv2.COLOR_BGR2GRAY)
hsv = np.zeros_like(frame1)
hsv[...,1] = 255

# saliency observation window
W = 10
w_counter = 1
s, m, roi, ids, obj_edges = sal_mask(frame1)
s_history = np.copy(s)
s_history = np.expand_dims(s_history, axis=2)

roi_history = []
roi_history.append(roi)
```

### Segmentation (enhanced by optical flow + smoothied to filter out unstable object recognition) + Saliency (smoothied from previous frames and filted out large variance)

```
### create processed frames and downsampled frames ###
k=1
while(1):
    ret, frame2 = cap.read()
    if ret == True:
      next = cv2.cvtColor(frame2,cv2.COLOR_BGR2GRAY)

      # get saliency (h,w) and masks, edges (h,w,#objects), saliency histories
      s, m, roi, ids, e = sal_mask(frame1)
      s2, m2, roi2, ids2, e2 = sal_mask(frame2)
      s_history = np.concatenate([s_history, np.expand_dims(s2, axis=2)], axis = 2)
      roi_history.append(roi2)

      if s_history.shape[2] < W: # if not observed 5 frames yet
        s_orig = np.copy(s)
      else: # if observed 5 frames
        # smooth saliency maps
        recent_saliens = s_history[:,:,-W:]
        recent_saliens_var = np.var(recent_saliens, axis = 2) #calculate recent saliency variance
        recent_saliens_var = normalize(recent_saliens_var)
        recent_saliens_var[recent_saliens_var>5e-05] = 0
        s_orig = np.mean(recent_saliens, axis = 2) #take mean of saliency maps
        np.putmask(s_orig, recent_saliens_var==0, 0)

        #filter out false positive object rois
        recent_roi = roi_history[-W:]
        recent_roi = [x for x in recent_roi if x.shape[0]>0]
        roi_update = recent_roi[0]
        for f in range(len(recent_roi)-1):
          # if roi_update.shape[0]>0 and recent_roi[f+1].shape[0]>0:
          roi_f = []
          dis = euclidean_distances(roi_update, recent_roi[f+1])
          n_obj1 = roi_update.shape[0]
          n_obj2 = recent_roi[f+1].shape[0]
          small_dim = np.argmin(dis.shape) #find fewer objects detected frame
          min_idx = dis.argmin(axis = 1-small_dim) #minimum value index for larger dimension

          if n_obj1 < n_obj2: #row is smaller dimension
              roi_f.append(recent_roi[f+1][min_idx,:])
          else:
              roi_f.append(roi_update[min_idx,:])
          roi_update = roi_f[0]
        
        # only need to filter roi2 based on roi_update when detect extra objects
        if roi.shape[0] > roi_update.shape[0]:
          dis = euclidean_distances(roi_update, roi)
          min_idx = dis.argmin(axis = 1-small_dim)
          e = e[:,:, min_idx]
          m = m[:,:, min_idx] 

      e_combined = np.sum(e, axis = 2)
      m_combined = np.sum(m, axis = 2)
      num_objs = m.shape[2]

      # optical flow
      flow = cv2.calcOpticalFlowFarneback(prvs,next, None, 0.5, 3, 15, 3, 5, 1.2, 0)
      mag, ang = cv2.cartToPolar(flow[...,0], flow[...,1])
      hsv[...,0] = ang*180/np.pi
      hsv[...,2] = cv2.normalize(mag,None,0,255,cv2.NORM_MINMAX)
      
      if num_objs > 0:
        # compute optical flow and saliency for each object within the mask
        o_opt = []
        o_s = []
        for o in range(num_objs):
          o_m = m[:,:,o]
          o_mag = np.mean(hsv[:,:,2] * o_m)
          o_opt.append(o_mag)
          o_s.append((s*o_m).mean())

        o_opt_n = normalize(np.array(o_opt).reshape(1,-1))
        o_s_n = normalize(np.array(o_s).reshape(1,-1))

        #highlight mask of e with high optical flow magnitude
        #highlight saliency of object mask 
        for o in range(num_objs):
          o_e = e[:,:,o]
          o_m = m[:,:,o]
          np.putmask(m_combined, o_m, 255 * o_opt_n[0][o])

      out = np.maximum(s_orig, m_combined)
      # out /= out.max()/1.0 
      out_60 = cv2.resize(out.astype('float64'), (60,60), interpolation=cv2.INTER_AREA)
      out_40 = cv2.resize(out.astype('float64'), (40,40), interpolation=cv2.INTER_AREA)
      out_20 = cv2.resize(out.astype('float64'), (20,20), interpolation=cv2.INTER_AREA)

      # plt.imshow(out,'gray')
      # plt.axis('off')
      # plt.savefig("%s/frame_%d.jpg" % (vid_path,k), bbox_inches='tight', pad_inches=0)

      plt.imshow(out_20, "gray")
      plt.axis("off")
      plt.savefig("%s/frame_%d_20.jpg" % (vid_path,k), bbox_inches='tight', pad_inches=0)

      plt.imshow(out_40, "gray")
      plt.axis("off")
      plt.savefig("%s/frame_%d_40.jpg" % (vid_path,k), bbox_inches='tight', pad_inches=0)
      
      plt.imshow(out_60, "gray")
      plt.axis("off")
      plt.savefig("%s/frame_%d_60.jpg" % (vid_path,k), bbox_inches='tight', pad_inches=0)

      k += 1
      prv = next
      frame1 = frame2

    else:
      break

cv2.destroyAllWindows()
cap.release()
```
